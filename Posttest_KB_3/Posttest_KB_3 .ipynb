# Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')

```

```python
plt.figure(figsize=(10, 6))  # Menentukan ukuran plot
plt.hist(data['Age'], bins=20, color='blue', edgecolor='black')  # Membuat histogram
plt.title('Distribusi Usia Pelanggan')  # Menambahkan judul
plt.xlabel('Age')  # Label sumbu x
plt.ylabel('Jumlah Pelanggan')  # Label sumbu y
plt.show()  # Menampilkan plot
```

## Kesimpulan

###  Histogram menunjukkan bahwa distribusi usia pelanggan cenderung memiliki bentuk yang mirip dengan distribusi normal atau Gaussian, dengan puncaknya berada di sekitar kelompok usia tertentu. Ini menunjukkan bahwa sebagian besar pelanggan berada pada kelompok usia tertentu, sementara jumlah pelanggan yang lebih muda atau lebih tua dari kelompok ini semakin berkurang. kita dapat melihat bahwa kelompok usia mayoritas pelanggan berada dalam rentang tertentu. Anda dapat mengidentifikasi rentang usia tersebut dengan melihat area puncak tertinggi di grafik.

# Pie chart

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
rating_counts = df['Rating'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Presentase Penilaian Berdasarkan Rating')
plt.axis('equal')  # Agar pie chart menjadi lingkaran sempurna
plt.show()

```

## Kesimpulan

### Mayoritas pelanggan memberikan rating tinggi, terutama rating 5. Ini menunjukkan bahwa sebagian besar produk dalam dataset mendapat penilaian positif dari pelanggan. 
### Rating 4 juga cukup tinggi, yang menunjukkan bahwa sebagian besar produk mendapat penilaian yang baik.
### Persentase produk yang mendapat rating 3 lebih kecil dibandingkan dengan rating yang lebih tinggi. Ini bisa dianggap sebagai kategori yang sedang atau standar.
### Rating 2 dan 1 memiliki presentase yang lebih kecil dibandingkan dengan rating yang lebih tinggi. Ini menunjukkan bahwa produk dengan rating rendah relatif jarang dalam dataset.

# Heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
heatmap_data = df.pivot_table(index='Rating', columns='Department Name', values='Clothing ID', aggfunc='count')
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt='g', linewidths=.5)
plt.title('Korelasi Antara Rating dan Department Name')
plt.xlabel('Department Name')
plt.ylabel('Rating')
plt.show()
```

## Kesimpulan

### Heatmap ini memungkinkan kita untuk melihat bagaimana penilaian (rating) berhubungan dengan departemen produk. Setiap baris pada heatmap mewakili tingkat rating (1 hingga 5), sementara setiap kolom mewakili department name.
### Dalam department name "Intimate," produk-produk dengan rating 4 dan 5 mendominasi, menunjukkan bahwa produk dalam kategori ini memiliki tingkat penilaian yang baik.
### Terdapat variasi penilaian di setiap department name, dengan semua tingkat rating (1-5) hadir dalam berbagai proporsi. Ini menunjukkan bahwa penilaian produk bervariasi tergantung pada departemen produk tersebut.
###  Beberapa department name seperti "Trend", "General", dan "Outerwear" juga memiliki variasi rating yang cukup tinggi. Ini menunjukkan bahwa penilaian untuk produk dalam department ini lebih merata.

# Grafik Batang

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Womens Clothing E-Commerce Reviews.csv')
class_counts = df['Class Name'].value_counts()
plt.figure(figsize=(12, 6))
class_counts.plot(kind='bar', color='skyblue')
plt.title('Sebaran Produk berdasarkan Class Name')
plt.xlabel('Class Name')
plt.ylabel('Jumlah')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

## Kesimpulan

### Dari diagram batang, kita dapat melihat bahwa kategori Dresses adalah yang paling dominan dalam dataset ini, diikuti oleh Knits dan Blouses. Ini menunjukkan bahwa produk-produk dalam kategori Dresses merupakan produk yang paling banyak diulas oleh pelanggan.
### Diagram batang juga memberikan wawasan tentang ragam produk yang tersedia dalam toko e-commerce tersebut. Kategori Dresses, Knits, dan Blouses memiliki jumlah produk yang signifikan, sedangkan kategori lainnya memiliki jumlah produk yang lebih rendah.
### Ada perbedaan signifikan dalam jumlah produk antara kategori-kategori. Hal ini dapat menjadi informasi yang berguna bagi pemilik bisnis atau pengecer untuk memahami mana kategori produk yang paling populer atau memiliki potensi penjualan tertinggi.
###  Karena beberapa kategori memiliki jumlah produk yang lebih sedikit, ini mungkin menunjukkan peluang untuk mengidentifikasi produk-produk unggulan dalam kategori tersebut yang dapat menjadi fokus pemasaran atau perbaikan.
